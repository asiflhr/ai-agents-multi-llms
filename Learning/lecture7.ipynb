```python
from langchain.agents.agent_types import AgentType
from langchain_experimental.agents.agent_toolkits import create_csv_agent
```

```python
import os
from dotenv import load_dotenv

load_dotenv()  # Load environment variables from .env file

# Corrected way to retrieve the API key
api_key = os.getenv("GOOGLE_API_KEY")

if not api_key:
    raise ValueError("GOOGLE_API_KEY not found. Make sure it's set in the .env file.")

os.environ["GOOGLE_API_KEY"] = api_key

from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash-latest")

from datetime import datetime
```

```python
agent = create_csv_agent(
    llm, 
    "titanic.csv", 
    verbose=True, 
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION, 
    allow_dangerous_code=True
)
```

```python
# LLM as a Judge

data_base = []

message_counter = 0

while True:
    user_input = input("User: ")
    message_counter += 1
    user_message = f"User Message {message_counter}: {user_input}"
    data_base.append(user_message)

    if (user_input.lower() == "exit") or (user_input.lower() == "quit") or (user_input.lower() == "bye") or (user_input.lower() == "end"):
        break

    result = agent.run(user_input)
    agent_message = f"Agent Message {message_counter}: {result}"
    data_base.append(agent_message)

    print(f"Agent: {result}")

llm_as_a_judge_prompt = f"""
You are a judge. You will output if the conversation is helpful or not based on the conversation and user_experience.
return "Helpful" or "Not Helpful" only no explanation needed.
conversation: {data_base}
"""
verdict = llm.invoke(llm_as_a_judge_prompt)
print(verdict.content)

print(data_base)
```
